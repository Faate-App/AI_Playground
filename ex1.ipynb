```python
import pandas as pd
import matplotlib.pyplot as plt
```

## Import data

```python
path = "home-data-for-ml-course/train.csv"
df = pd.read_csv(path)
df.head(n=50)
```

 ## Data analysis

```python
features = ['BedroomAbvGr', 'PoolArea', 'GarageArea', 'FullBath',
                        'YearBuilt', 'GarageCars', 'TotRmsAbvGrd']
X = df[features]
X = X.dropna(axis=0)
X.head()
```

```python
y = df["SalePrice"]
```

```python
from sklearn.tree import DecisionTreeRegressor
from sklearn.metrics import mean_absolute_error
from sklearn.model_selection import train_test_split
```

```python
train_X, val_X, train_y, val_y = train_test_split(X, y,random_state = 0)
```

```python
forest_model = DecisionTreeRegressor(random_state=1)
forest_model.fit(train_X, train_y)
melb_preds = forest_model.predict(val_X)
```

```python
print("Making predictions for the following 5 houses:")
print("The predictions are")
print(forest_model.predict(val_X.head(n=10)))
print(val_y.head(n=10))

# Error calculation
```
